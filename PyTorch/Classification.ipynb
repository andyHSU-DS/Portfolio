```python
import torch
import torch.nn.functional as F
from torch.autograd import Variable
import matplotlib.pyplot as plt
```

```python
n_data = torch.ones(100, 2)
x0 = torch.normal(2*n_data, 1)      # class0 x data (tensor), shape=(100, 2)#torch.normal(mean,std)
y0 = torch.zeros(100)               # class0 y data (tensor), shape=(100, 1)
x1 = torch.normal(-2*n_data, 1)     # class1 x data (tensor), shape=(100, 2)
y1 = torch.ones(100)                # class1 y data (tensor), shape=(100, 1)
x = torch.cat((x0, x1), 0).type(torch.FloatTensor)  # shape (200, 2) FloatTensor = 32-bit floating
y = torch.cat((y0, y1), ).type(torch.LongTensor)    # shape (200,) LongTensor = 64-bit integer

x=Variable(x)
y=Variable(y)
```

```python
plt.scatter(x.data.numpy()[:,0],x.data.numpy()[:,1],c=y.data.numpy(), s=100, lw=0, cmap='RdYlGn')
```

```python
class Net(torch.nn.Module):#繼承
    def __init__(self,n_feature,n_hidden,n_output):#搭網路要的訊息
        super(Net,self).__init__()
        self.hidden = torch.nn.Linear(n_feature,n_hidden)
        self.predict = torch.nn.Linear(n_hidden,n_output)
        
    def forward(self,x):#前項傳遞的過程
        x = F.relu(self.hidden(x))#activation function
        x = self.predict(x)
        return x
```

```python
net = Net(2,10,2)#2(x,y軸數值)個特徵，10個神經元，2個輸出ex:[0,1],[1,0]三分類的話[1,0,0][0,1,0][0,0,1]
print(net)
```

```python
optimizer = torch.optim.SGD(net.parameters(),lr=0.02)
loss_func = torch.nn.CrossEntropyLoss()#針對多分類，計算softmax
```

```python
for t in range(100):
    out = net(x)#因為出來的結果會是[-2,-3]#並不是分類，要再依靠softmax將它轉成機率
    #print(out)
    #print(torch.max(out,1))
    loss = loss_func(out,y)
    
    optimizer.zero_grad()#將梯度都設為0
    loss.backward()#微分
    optimizer.step()#根據lr調整參數
    if t % 2 == 0:
            # plot and show learning process
            plt.cla()
            prediction = torch.max(out, 1)[1]
            pred_y = prediction.data.numpy()
            target_y = y.data.numpy()
            plt.scatter(x.data.numpy()[:, 0], x.data.numpy()[:, 1], c=pred_y, s=100, lw=0, cmap='RdYlGn')
            accuracy = float((pred_y == target_y).astype(int).sum()) / float(target_y.size)
            plt.text(1.5, -4, 'Accuracy=%.2f' % accuracy, fontdict={'size': 20, 'color':  'red'})
            plt.pause(0.1)


plt.ioff()
plt.show()
```
